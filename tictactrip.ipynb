```python
import numpy as np
import pandas as pd

BASE_DIR = "./data/"

tickets_df = pd.read_csv(BASE_DIR+'ticket_data.csv')
tickets_df.head()
```

```python
cities_df = pd.read_csv(BASE_DIR+'cities.csv')
cities_df.head()
```

```python
stations_df = pd.read_csv(BASE_DIR+'stations.csv')
stations_df.head()
```

```python
providers_df = pd.read_csv(BASE_DIR+'providers.csv')
providers_df.head()
```

Pour savoir le nombre des compagnies, stations, villes..., j'ai utilisé la fonction nunique.

```python
tickets_df.nunique()
```

Pour extraire les infos demandées, j'avais besoin de calculer la duration en heures de chaque trajet. Ensuite, j'ai ajouté une nouvelle colomne de duration au tableau "tickets_df"

```python
tickets_df['departure_ts'] = pd.to_datetime(tickets_df['departure_ts'])
tickets_df['arrival_ts'] = pd.to_datetime(tickets_df['arrival_ts'])
tickets_df['duration_in_hours'] = (tickets_df['arrival_ts'] - tickets_df['departure_ts'])/np.timedelta64(1,'h')
tickets_df.head()
```

J'ai créé un autre dataframe ("infoTrajet_df") pour afficher les informations demandées. Avec groupby, j'ai regroupé les trajets qui ont les mêmes villes d'origine et de destination.

```python
infoTrajet_df = tickets_df.groupby(['o_city', 'd_city']).size().reset_index(name='nb_tickets')
print (infoTrajet_df)
```

Ensuite, j'ai calculé le prix et duration min, max, moyen pour chaque trajet et puis, je les ai ajouté au dataframe infoTrajet_df

```python
averagePrice_column = tickets_df.groupby(['o_city', 'd_city']).mean()['price_in_cents']
minPrice_column = tickets_df.groupby(['o_city', 'd_city']).min()['price_in_cents']
maxPrice_column = tickets_df.groupby(['o_city', 'd_city']).max()['price_in_cents']

averageDuration_column = tickets_df.groupby(['o_city', 'd_city']).mean()['duration_in_hours']
minDuration_column = tickets_df.groupby(['o_city', 'd_city']).min()['duration_in_hours']
maxDuration_column = tickets_df.groupby(['o_city', 'd_city']).max()['duration_in_hours']


infoTrajet_df = infoTrajet_df.set_index(['o_city', 'd_city','nb_tickets'])

infoTrajet_df['average_price_in_cents'] = averagePrice_column
infoTrajet_df['min_price_in_cents'] = minPrice_column
infoTrajet_df['max_price_in_cents'] = maxPrice_column

infoTrajet_df['average_duration_in_hours'] = averageDuration_column
infoTrajet_df['min_duration_in_hours'] = minDuration_column
infoTrajet_df['max_duration_in_hours'] = maxDuration_column


infoTrajet_df = infoTrajet_df.reset_index()
infoTrajet_df.head()
```

Le dataframe ci-dessus, collecte les informations demandées selon le trajet. Nous pouvons noter que pour un même trajet, nous pouvons trouver 2 tickets avec un difference de duration et prix enorme.


Pour le deuxieme exercise, les infos demandées seront calculées selon la distance de chaque trajet. Pour ce faire, j'ai croisé les infos entres les differents tableaux pour recuperer les coordonés de chaque ville. J'ai créé un nouveau dataframe pour garder toutes ces informations.

```python
new_df = pd.merge(left=tickets_df, right=cities_df, left_on='o_city', right_on='id',how='inner')
new_df.head()
```

```python
new_df = pd.merge(left=new_df, right=cities_df, left_on='d_city', right_on='id',how='inner')
new_df.head()
```

Avec les coordonées, j'ai calculé les distances de chaque trajet avec  la function geodesic de la librairie geopy. Puis, j'ai ajouté une nouvelle colonne de distance.

```python
from geopy.distance import geodesic 
def distance_calc (row):
    start = (row['latitude_x'], row['longitude_x'])
    stop = (row['latitude_y'], row['longitude_y'])

    return geodesic(start, stop).km

new_df['distance'] = new_df.apply (lambda row: distance_calc(row),axis=1)
new_df.head()
```

Ensuite, j'ai utilisé au nouveau la fonction merge pour croiser les infos des providers et recuperer le type de transport de chaque trajet.

```python
new_df = pd.merge(left=new_df, right=providers_df, left_on='company', right_on='company_id',how='inner')
new_df.head()
```

Après, j'ai affiché les colonnes de ce nouveau dataframe pour eliminer celles qui n'apportent pas d'information utile pour cet exercise.

```python
new_df.columns
```

```python
new_df = new_df.drop(columns=['id_y','local_name_x', 'unique_name_x', 'latitude_x', 'longitude_x',
       'population_x', 'id_x', 'local_name_y', 'unique_name_y', 'latitude_y',
       'longitude_y', 'population_y', 'id_y', 'company_id',
       'provider_id', 'name', 'fullname', 'has_wifi', 'has_plug',
       'has_adjustable_seats', 'has_bicycle'])
new_df.head()
```

J'ai trouvé la distance maximale pour savoir les ranges dans lequelles je devais diviser cette variable et puis calculer les infos demandées.

```python
new_df.transport_type.isna().sum()
```

```python
new_df['distance'].max()
```

```python
bins = pd.cut(new_df['distance'], [0, 200, 800, 2000])
info_df = new_df.groupby(bins)['distance'].agg(['count'])
info_df.head()
```

Finalement, j'ai utilisé la fonction groupby pour regrouper les trajet selon les ranges definies précédement  et selon le type de transport, en calculant les prix et durations moyens.

```python
bins = [0, 200, 800, 2000]
priceInfo_df = new_df.groupby([pd.cut(new_df.distance, bins), 'transport_type']).mean()['price_in_cents']
priceInfo_df.unstack()
```

```python
durationInfo_df = new_df.groupby([pd.cut(new_df.distance, bins), 'transport_type']).mean()['duration_in_hours']
durationInfo_df.unstack()
```
